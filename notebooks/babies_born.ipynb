```python
import pandas as pd
import matplotlib.pyplot as plt
data_path_ludnosc = "../ludnosc.csv"
data_path_gminy = "../gminy.csv"
df_gminy = pd.read_csv(data_path_gminy, sep=";", decimal=',', low_memory=False)
```

```python
babies_born_df = df_gminy[(df_gminy.Zmienna == "urodzenia żywe na 1000 ludności")]
babies_born_df = babies_born_df[(babies_born_df.Nazwa.str.isupper() == True)]
babies_born_df = babies_born_df.drop(['Atrybut','Unnamed: 8','Jednostka miary','Kategoria','Zmienna','Kod'], axis=1)
display(babies_born_df)
```

## Liczba żywych urodzeń na 1000 ludności w Polsce na przestrzeni lat

```python
births_over_years = babies_born_df[(babies_born_df.Nazwa == "POLSKA")]

display(births_over_years)
plt.figure()
plt.figure(figsize=(10,5))
plt.title('Liczba żywych urodzeń na 1000 ludności w Polsce na przestrzeni lat', size=25)
plt.xlabel('Rok')
plt.ylabel('Liczba żywych urodzeń')
plt.xlim(2002,2020)
plt.ylim(8,12)
plt.xticks(births_over_years['Rok'])
plt.plot(births_over_years['Rok'],births_over_years["Wartosc"], color='orange')
```

## Gdzie i kiedy urodziło się najmniej i najwięcej dzieci?


```python
min_max_births = babies_born_df.sort_values('Wartosc', ascending=False)
display(min_max_births.head(1).style.set_caption("Największa ilość urodzeń:"))
display(min_max_births.tail(1).style.set_caption("Najmniejsza ilość urodzeń:"))
```
