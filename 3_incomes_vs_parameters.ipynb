```python
df = pd.DataFrame()
df["Kraje"] = Kraje.tolist()
df["Dochody"] = dochody.tolist()
df["wydatki kultura"] = wydatki.tolist()
df["udzial kultury"] = df["wydatki kultura"]/df["Dochody"]
df.sort_values(by = ["udzial kultury"])
import matplotlib.pyplot as plt
x = df["Dochody"]
y= df["wydatki kultura"]
plt.scatter(x,y)
plt.show()
```
